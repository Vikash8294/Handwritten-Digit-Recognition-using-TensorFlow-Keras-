```python
# Import necessary libraries for building and evaluating the neural network
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

```

```python
(x_train,y_train),(x_test, y_test) = keras.datasets.mnist.load_data();
```

```python
# Normalize the original training and testing data by dividing by 255
x_train1 = x_train/255
x_test1 = x_test/255

```

```python
x_train1[1]
```

```python
# Define a third model using a Flatten layer as the first layer
model= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam', loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
model.fit(x_train1,y_train, epochs = 5)
```

```python
model.evaluate(x_test1,y_test)
```

```python
y_pred = model.predict(x_test1)
y_pred[1]
```

```python
np.argmax(y_pred[1])
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
y_pred_label[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label)
cm
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')

    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename
```

```python
from IPython.display import Image

try:
    filename = take_photo()
    print('Saved to {}'.format(filename))
    # Show the image which was just taken.
    display(Image(filename))
except Exception as err:
    # Errors will be thrown if the user does not have a webcam
    # or if they do not grant the page permission to access it.
    print(str(err))
```

```python
# Import necessary libraries
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# 'filename' is the photo you just captured using take_photo()
user_test = filename

# 1️⃣ Open the captured image
col = Image.open(user_test)

# 2️⃣ Convert the image to grayscale
# Grayscale means only shades of black & white (no color)
gray = col.convert('L')

# 3️⃣ Convert the grayscale image to pure black and white
# Pixels below 100 become black (0), others become white (255)
bw = gray.point(lambda x: 0 if x < 100 else 255, '1')

# 4️⃣ Save the black-and-white version
bw.save("bw_image.jpg")

# 5️⃣ Read the saved image using OpenCV in grayscale mode
img_array = cv2.imread("bw_image.jpg", cv2.IMREAD_GRAYSCALE)

# 6️⃣ Invert colors — MNIST digits are white text on black background,
# so we flip the colors to match the dataset format.
img_array = cv2.bitwise_not(img_array)

# 7️⃣ Print the shape (height × width)
print("Original image size:", img_array.shape)

# 8️⃣ Display the black-and-white inverted image
plt.imshow(img_array, cmap=plt.cm.binary)
plt.title("Processed User Image")
plt.show()

# 9️⃣ Resize the image to 28×28 pixels to match MNIST input size
img_size = 28
new_array = cv2.resize(img_array, (img_size, img_size))

plt.imshow(new_array, cmap=plt.cm.binary)
plt.title("Resized 28x28 Image")
plt.show()

# 🔟 Normalize pixel values between 0 and 1 for model input
user_test = tf.keras.utils.normalize(new_array, axis=1)

# 1️⃣1️⃣ Reshape the image for prediction
# Model expects a 4D array: [batch_size, height, width, channels], but our model has a Flatten layer
# so we need to reshape to [batch_size, height, width] or [1, 28, 28]
user_test = np.expand_dims(user_test, axis=0)

# 1️⃣2️⃣ Predict using the trained model
predicted = model.predict(user_test)

# 1️⃣3️⃣ Display probability distribution for each digit (0–9)
for i in range(10):
    print("Probability of", i, ":", predicted[0][i])

# 1️⃣4️⃣ Print the final predicted digit
print("\n✅ The Predicted Value is:", np.argmax(predicted[0]))
```
